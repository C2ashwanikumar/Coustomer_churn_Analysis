```python
# WE converter Senior Citizen Value from 0 and 1 to YES or NO
```

```python
def conv(value):
    if value == 1:
       return "Yes"
    else:
       return "No"
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
plt.pie(gb['Churn'])
labels = gb['Churn'].index  
sizes = gb['Churn'].values  

# Create a pie chart
plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%',  
    startangle=90,      # Start from the top
    colors=['skyblue', 'orange'],  
    explode=[0, 0]   
)

plt.title('Churn Distribution') 
plt.show()
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Define the number of rows and columns for subplots
n_cols = 3  # Set the number of columns for subplots
n_rows = len(columns) // n_cols + (len(columns) % n_cols > 0)  # Calculate number of rows

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop over the columns and create a countplot for each
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Churn by {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove any unused axes (in case number of subplots is greater than the number of columns)
for j in range(i+1, len(axes)):
    axes[j].axis('off')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

#1. Phone Service:

# Customers with phone service are more likely to stay (non-churned) than leave.

# Very few customers, whether churned or non-churned, lack phone service.

#2. Multiple Lines:

# Both customers with multiple lines and without are less likely to churn.

# Customers without phone service remain a small group.

# 3. Internet Service:

# DSL and Fiber Optic services see a higher number of non-churned customers compared to churned.

# Customers without internet service are less likely to churn.

#4. Online Security:

# More customers with online security tend to stay.

# Those without online security have a higher churn rate, but still show a majority of non-churned customers.

#5. Online Backup:

# Similar to online security, more customers with online backup are non-churned.

#6. Device Protection:

# Non-churned customers are more prevalent among those with device protection.

# 7.Tech Support:

# Customers with tech support services are more likely to stay.

# 8.Streaming TV and Movies:

# Both services have a higher number of non-churned customers.
